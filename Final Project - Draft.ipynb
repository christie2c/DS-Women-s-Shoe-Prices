```python
#BRAND

#Making copy of df with relevant columns only:
df = df0[['brand', 'prices.amountMax', 'prices.amountMin', 'prices.merchant']]
df.brand = df.brand.str.lower()
df['prices.merchant'] = df['prices.merchant'].str.lower()
df["Avgprice"] = (df["prices.amountMax"] + df["prices.amountMin"])/2
df.head()
```

```python
df.brand.value_counts()
```

```python
print(df.brand.unique())
print(df.brand.nunique())
```

```python
#calculate the number of shoes per brand, sort descending
brand_counts = df.brand.value_counts().sort_values(ascending=False)
top_brands = brand_counts.head(50)
top_brands
```

```python
#BRAND Graph (1) - Top Brands in this Dataset
total = brand_counts.cumsum()
top_brands = brand_counts[total <= 7500]
print(top_brands)
```

```python
plt.bar(top_brands.index, top_brands.values)
plt.xlabel('Brand')
plt.ylabel('Number of Shoes')
plt.title('10 Brands contain the Top 75% pairs of shoes (7500) in this dataset')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#BRAND (2) - Average Price Per Brand

#making the binned by brand datafram
df_bin1 = df.groupby('brand')
avg_price = df_bin1['Avgprice'].mean().sort_values(ascending=False)
print(avg_price)
print(type(df_bin1))
print(type(avg_price))

#Note: df_bin type = DataFrameGroupBy (?binned values cannot become a new DataFrame)
#Note: ave_priceMax = Series (column extracted from the temp binned DataFrame)
```

```python
#BRAND Graph (2) Top 20 Brands with the highest average prices
avg_price.head(20)

plt.bar(avg_price.head(20).index, avg_price.head(20).values)
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.title('Average Price for Each Brand for the top 20 most expensive brands')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_price.head()
```

```python
#Brands with the largest sales  (largest Max = Min)
df["Maxmin"] = df["prices.amountMax"] - df["prices.amountMin"]
```

```python
df_bin2 = df.groupby("brand")
largestdiscount = df_bin2['Maxmin'].mean().sort_values(ascending = False)
print(type(largestdiscount))
largestdiscount
```

```python
#BRAND Graph (3): Top 20 Brands with the largest discounts

largestdiscount.head(20)
```

```python
plt.bar(largestdiscount.head(20).index, largestdiscount.head(20).values)
plt.xlabel('Brands')
plt.ylabel("Discount (Max price - min price)")
plt.title("Brands offering the largest discounts")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#(4) plot scatter plot using seaborn with y = maxmin and x = avgprice, hue = binned brands by avg price
df.Avgprice.sort_values(ascending = False).median()
```

```python
median = 56.875

#Bin again by average price
df_bin1 = df.groupby('brand')
avg_price = df_bin1['Avgprice'].mean().sort_values(ascending=False)
avg_price


#Map price category to avgprice acc to bands bin to sort brands into into high price (True) or low price (False)
def price_category(avg_price):
    return avg_price > median
df['PriceHigh'] = df['Avgprice'].map(price_category)

df_sns = df[df['prices.amountMax'] != df['prices.amountMin']]
df_sns
```

```python
sns.scatterplot(x ="Avgprice", y ="Maxmin", data=df_sns, hue ="PriceHigh")
plt.xlabel('Average Price per Brand')
plt.ylabel('Size of Discount (Max price - Min price)')

#?? should i remove the ones where min = max?
```
